```python
# importando as bibliotecas
import pandas as pd
import numpy as np
import io
import plotly.express as px
import folium

from google.colab import files
```

```python
# importando o arquivo do computador para a nuvem
uploaded_files = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded_files['AB_NYC_2019.csv']))
```

```python
# 1 - Qual o valor médio do aluguél na Cidade de Nova York?
df.loc[: , 'price'].mean()
```

```python
# 2 - Quais os nomes das regiões que existem na cidade de Nova York? R: Contar as regiões uma única vez
np.unique(df.loc[:, 'neighbourhood_group'])
```

```python
# 3- Qual o valor do aluguél diário mais caro da cidade de Nova York? R: Pegar a coluna price e selecionar o maior valor
df.loc[:, 'price'].max()
```

```python
# 4 - Quais são as categorias de imóveis que estão cadastradas na base de dados da cidade de Nova York? R: Contar as categorias existentes uma única vez
np.unique(df.loc[:, 'room_type'])
```

```python
# 5 - Quantos usuários (Hosts) únicos cadastros dentro da base de dados da cidade de Nova York?
len(np.unique(df.loc[:, 'host_id'])) #uso a função len() para contar o tamanho da minha lista
```

```python
# 6 -  Como é a variação de preços dos imóveis em Nova York?
np.std(df.loc[:, 'price'])
```

```python
# 7 - Existem mais imóveis baratos ou caros? 
px.histogram(df.loc[df.loc[:, 'price'] < 1000, :], 'price', nbins=300)
# R: Sendo a média de preços de U$$152,00, é possivel observar que que ha mais imóveis concentrados abaixo da média. Portanto, existem mais imóveis mais baratos do que caros.
```

```python
# 8 - Como é a distribuição do número de Reviews? Existem imóveis com muitos e outro com poucos reviews?
px.histogram(df.loc[:, 'number_of_reviews'], 'number_of_reviews', nbins=300)
# R: A maioria dos imóveis possui até 4 reviews
```

```python
# 9 - Qual o valor do alguél (diária) mais caro de cada de cada região da base de dados da cidade de Nova York, apenas para imóvies disponíveis para alugar?
df = df.loc[df['availability_365'] > 0 , :] # filtrando apenas os imóveis que estão diponíveis para alugar
data_plot = df.loc[:, ['neighbourhood_group', 'price']].groupby('neighbourhood_group').max().reset_index()

px.bar(data_plot, x='neighbourhood_group', y='price')
```

```python
#10 - Conseguimos saber onde estão localizados os imóveis com o alugél mais mais caro, na cidade de Nova York, apenas para imóveis para alugar? R: Fazer um mapa mostrando os alugeis mais caros por região
data_plot = df.loc[:, ['neighbourhood_group', 'price', 'latitude', 'longitude']].groupby('neighbourhood_group').max().reset_index()
map = folium.Map()
for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'],
                 location_info['longitude']],
                 popup=location_info['neighbourhood_group']).add_to(map)

map

```

```python
# 11 - Conseguimos saber onde estão localizados os imóveis pelo seu tipo, apenas para imóveis disponíveis para alugar
data_plot = df.loc[: , ['neighbourhood_group', 'room_type', 'latitude', 'longitude']].sample(50)

# criando uma coluna de cores
data_plot.loc[:, 'color'] = 'NA'

data_plot.loc[data_plot['room_type'] == 'Private room', 'color'] = 'darkgreen'
data_plot.loc[data_plot['room_type'] == 'Entire home/apt', 'color'] = 'darkred'
data_plot.loc[data_plot['room_type'] == 'Shared room', 'color'] = 'purple'

map = folium.Map()

for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'], location_info['longitude']],
                popup=location_info['neighbourhood_group'],
                icon=folium.Icon(color=location_info['color'])).add_to(map)

map
```
